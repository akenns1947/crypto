```python
import pandas as pd
import os
```

```python
#load VES and VEF
VEF = pd.read_csv('../temporary/.localbtcVEF.csv', names = ['date', 'price', 'amount'])
VES = pd.read_csv('../temporary/.localbtcVES.csv', names = ['date', 'price', 'amount'])
```

```python
VEF = VEF.drop_duplicates()
VES = VES.drop_duplicates()

VEF.insert(1, 'tid', range(len(VEF)))
VES.insert(1, 'tid', range(len(VEF), len(VEF)+len(VES)))

VEF['Currency'] = 'VEF'
VES['Currency'] = 'VES'
```
