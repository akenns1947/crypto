```python
import pandas as pd
from datetime import datetime
import os
```

```python
trades = pd.read_csv('../temporary/trades_combined.csv')
trades
```

```python
trades['date'] = pd.to_datetime(trades['date'], unit = 's')
trades
```

```python
trades = trades.sort_values(by='date')
trades = trades.drop_duplicates(subset = ['tid'])
```

```python
trades
```

```python
#count duplicates
from collections import Counter
count = Counter(trades['tid'])

dupes = sum(v for k, v in count.items() if v > 1)
dupes
```
